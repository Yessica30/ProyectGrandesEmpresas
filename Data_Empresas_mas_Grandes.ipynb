Descripcion de los 5 principales macrosectores mas utilizados por las empresas para el año 2021:

**COMERCIO**: Hace parte del sector terciario de la economía, e incluye comercio al por mayor, minorista, centros comerciales, cámaras de comercio, las plazas de mercado y, en general, todos aquellos que se relacionan con la actividad de comercio a nivel nacional o internacional.

**SERVICIOS**: El sector servicios o sector terciario es el sector económico que engloba las actividades relacionadas con los servicios no productores o transformadores de bienes materiales. Generan servicios que se ofrecen para satisfacer las necesidades de cualquier población en el mundo.

**MANUFACTURA**: La industria manufacturera se define como la transformación mecánica o química de sustancias orgánicas e inorgánicas en productos nuevos, elaborados bien sea a mano o con maquinaria.

**CONSTRUCCION** se encuentra en el sector secundario, son empresas que centran su actividad productiva transformando bienes en otros más útiles para su uso, agregándoles valor. En este grupo se encuentran las empresas industriales y de construcción.

**AGROPECUARIO**:  se dedica a la obtención de materia prima de origen vegetal y animal, destinada a la industria alimentaria o manufacturera.

el sector comercio es el más prevalente entre las empresas más grandes de Colombia en 2021. Esto sugiere que una proporción significativa de las principales empresas del país se dedica a actividades comerciales

La gráfica también sugiere que otros sectores, aunque importantes, no tienen la misma presencia entre las empresas más grandes. Esto podría señalar oportunidades de crecimiento y desarrollo en sectores como manufactura, servicios y tecnología.
